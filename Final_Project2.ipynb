   In recent years, the audiobook industry has emerged as one of the fastest-growing         segments of the digital content market. The Audible Dataset provides valuable insights into this evolving market by capturing key information about audiobooks released between 1998 and 2025 (including pre-planned releases). It contains attributes such as:

####  <span style="color:green"> Capitalize the first letter of each column name </span>

```python
Df.columns = Df.columns.str.title()
```

```python
Df
```

####  <span style="color:green"> Removing unwanted words like 'Writtenby:' and 'Narratedby:' from columns </span>


```python
Df['Author'] = Df['Author'].str.replace('Writtenby:', '', regex=False)
Df['Narrator'] = Df['Narrator'].str.replace('Narratedby:', '', regex=False)
```

```python
Df
```

#### <span style="color:green"> Adding space between the first and second name </span>


```python
Df['Author'] = Df['Author'].str.replace(r'([a-z])([A-Z])', r'\1 \2', regex=True)
Df['Narrator'] = Df['Narrator'].str.replace(r'([a-z])([A-Z])', r'\1 \2', regex=True)
Df
```

####  <span style="color:green">#removing 'and' from Time cloumn  </span>


```python

Df['Time'] = Df['Time'].str.replace('and', '', regex=False)
Df
```

####  <span style="color:green">Splitting the column Stars into two cloumns as Star rating and Total rating  </span>


```python

Df['Star_Rating'] = Df['Stars'].str.extract(r'(\d+\.?\d*)')



Df['Total_Ratings'] = Df['Stars'].str.extract(r'(\d+)\s*ratings')
```

```python
Df
```

```python
#Droping Star column
Df.drop(columns=['Stars'], inplace=True)
```

```python
Df
```

####  <span style="color:green"> Renaming the Column 'Name' </span>


```python
Df.rename(columns={'Name': 'Title'}, inplace=True)
print(Df)
```

```python
 print(Df.isnull().sum())
```

```python
#Changing the datatype
Df['Star_Rating'] =  Df['Star_Rating'].astype(float)
print(Df.dtypes)
```

####  <span style="color:green">Filling missing values in Star_Rating  </span>


```python


# Filling missing with author-wise mean, then global mean
Df['Star_Rating'] = Df.groupby('Author')['Star_Rating'].transform(lambda x: x.fillna(x.mean()))
Df['Star_Rating'] = Df['Star_Rating'].fillna(Df['Star_Rating'].mean())

print("After:", Df['Star_Rating'].isnull().sum())
```

```python
Df
```

```python
# Rounding the decimal places
Df['Star_Rating_rounded'] = Df['Star_Rating'].round(1)
Df
```

####  <span style="color:green">Filling missing values in Total_Ratings  </span>

```python
#Changing the datatype
Df['Total_Ratings'] =  Df['Total_Ratings'].astype(float)
print(Df.dtypes)
```

```python
# Fill missing with author-wise mean, then global mean
Df['Total_Ratings'] = Df.groupby('Author')['Total_Ratings'].transform(lambda x: x.fillna(x.mean()))
Df['Total_Ratings']  = Df['Total_Ratings'].fillna(Df['Total_Ratings'].mean())

# Verify
print("After:", Df['Total_Ratings'].isnull().sum())
```

```python
Df
```
